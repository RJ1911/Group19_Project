```python
import pandas as pd
```

```python
df1 = pd.read_csv('combined_data_clean.csv', encoding='latin1')
df2 = pd.read_csv('UpdatedResumeDataSet.csv')
```

```python
import re
import nltk
from bs4 import BeautifulSoup

def cleanResume2(html_data):
    # Remove HTML tags
    soup = BeautifulSoup(html_data)
    text = soup.get_text()

    # Remove unwanted characters
    text = re.sub(r'[^\w\s]', '', text)

    # Normalize text
    text = text.lower()
    text = nltk.PorterStemmer().stem(text)

    # Remove stop words
    stopwords = nltk.corpus.stopwords.words('english')
    text = ' '.join([word for word in text.split() if word not in stopwords])

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Handling contractions
    text = nltk.word_tokenize(text)
    text = [nltk.WordNetLemmatizer().lemmatize(word) for word in text]
    text = ' '.join(text)

    # Removing Non−Textual Elements
    text = re.sub(r'\[[^\]]+\]', '', text)
    return text
```

```python
df2['resume'] = df2['resume'].apply(cleanResume2)
```

```python


# Combine the DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("final_resume_dataset.csv", index=False)



```
